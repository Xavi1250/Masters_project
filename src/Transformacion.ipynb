```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("/Users/xavi/Documents/Ironhack/Projects/Masters_project/MBAs_Spain.csv")
```

```python
df["Tiempo"]=df["Tiempo"].replace({"1 year": "12", "1½ year":"18"})
df = df.rename(columns={"Tiempo": "Tiempo (months)", "Precio":"Precio (por año)"})
```

```python
for i in df["Tiempo (months)"]:
    if "months" in i:    
        df["Tiempo (months)"] = df["Tiempo (months)"].str.replace("months", "")

df = df.sort_values(by="Precio (por año)", ascending=False)
df.reset_index(drop=True)
```

```python
df1 =df.groupby("Presencial")["Precio (por año)"].mean().reset_index()
df1
```

```python
sns.barplot(x="Presencial", y="Precio (por año)", data=df1)
```
